```python
import cv2
import mediapipe as mp
import pandas as pd
import numpy as np
from IPython.display import display, clear_output

# Initialize Mediapipe Pose model and drawing utilities
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# Function to extract landmarks and create dataset
def create_pose_dataset(video_path, output_csv="body_landmarks_dataset.csv"):
    data = []  # List to store landmark data for each frame

    # Open video file
    cap = cv2.VideoCapture(r"C:\Users\Pooja\Desktop\FYP\jp\squats.mp4")

    # Initialize Mediapipe Pose
    with mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
        frame_num = 0
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break  # End of video

            # Convert frame to RGB for MediaPipe processing
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = pose.process(image)

            # Collect landmarks if detected
            if results.pose_landmarks:
                landmarks = results.pose_landmarks.landmark
                frame_data = {'frame': frame_num}

                # Save x, y, z coordinates and visibility for each landmark
                for idx, landmark in enumerate(landmarks):
                    frame_data[f"x_{idx}"] = landmark.x
                    frame_data[f"y_{idx}"] = landmark.y
                    frame_data[f"z_{idx}"] = landmark.z
                    frame_data[f"visibility_{idx}"] = landmark.visibility

                # Append frame data to list
                data.append(frame_data)
                frame_num += 1

                # Display the frame in the notebook
                annotated_frame = frame.copy()
                mp_drawing.draw_landmarks(annotated_frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
                display_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)
                clear_output(wait=True)
                display(display_frame)

    # Release video capture
    cap.release()

    # Convert data list to DataFrame and save as CSV
    df = pd.DataFrame(data)
    df.to_csv(output_csv, index=False)
    print(f"Dataset created and saved as '{output_csv}'")

# Run the function on your video file
create_pose_dataset(r"C:\Users\Pooja\Desktop\FYP\jp\squats.mp4")  # Replace with your video file path
```

```python
import torch
import cv2

# Load trained YOLO model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='best.pt', force_reload=True)

# Start video capture from webcam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform pose detection
    results = model(frame)

    # Parse results
    for detection in results.xyxy[0]:  # x1, y1, x2, y2, confidence, class
        x1, y1, x2, y2, conf, cls = map(int, detection[:6])

        # Label and color for correct/incorrect pose
        label = 'Correct Pose' if cls == 0 else 'Incorrect Pose'  # Assuming class 0 is correct pose
        color = (0, 255, 0) if cls == 0 else (0, 0, 255)

        # Draw bounding box and label on frame
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, f'{label} ({conf:.2f})', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # Display frame with detections
    cv2.imshow('Pose Detection', frame)

    # Press 'q' to exit the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
